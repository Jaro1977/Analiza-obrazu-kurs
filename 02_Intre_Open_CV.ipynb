```python
from google.colab.patches import cv2_imshow
import imutils
import numpy as np
import cv2
```

```python
cv2.__version__
```

```python
img = cv2.imread(filename = 'bear.jpg')
cv2_imshow(img)
```

```python
img_resized = imutils.resize(image = img, height = 300)
cv2_imshow(img_resized)
```

```python
img_resized = imutils.resize(image = img, width = 300)
cv2_imshow(img_resized)
```

```python
height, width = img.shape[:2]

M = cv2.getRotationMatrix2D(center = (height // 2, width // 2), angle =45 , scale =.8 )
cv2_imshow(cv2.warpAffine(src = img, M=M, dsize = (width,height)))
```
